```python
# 5.1 Plot the minimum running time obtained from the three experiments (for each case),
# i.e., 1 plot, x-axis: number of cores, y-axis: running time in seconds..
# with constant r = 1 i.e STRONG SCALING

import matplotlib.pyplot as plt
p = [1, 2, 4, 8, 16, 24, 32]
min_time = [
    0.0375154,
    0.0462122,
    0.0239887,
    0.0288063,
    0.026024,
    0.0268065,
    0.0419064
]
plt.plot(p, min_time, label = "r = 1 (strong scaling)")
plt.xlabel('Number of cores')
plt.ylabel('running time (s)')
plt.title('Minimum running time vs Numbers of cores for r = 1 (strong scaling)')
plt.legend()
plt.show()
```

```python
# 5.2 Plot the minimum running time obtained from the three experiments (for each case),
# i.e., 1 plot, x-axis: number of cores, y-axis: running time in seconds..
# with constant r = p i.e WEAK SCALING

import matplotlib.pyplot as plt
p = [1, 2, 4, 8, 16, 24, 32]
min_time = [
    0.0375121,
    0.131669,
    0.0775685,
    0.0427748,
    0.0242519,
    0.062618,
    0.0947295
]
plt.plot(p, min_time, label = "r = p (weak scaling)")
plt.xlabel('Number of cores (p)')
plt.ylabel('running time (s)')
plt.title('Minimum running time vs Numbers of cores for r = p (weak scaling)')
plt.legend()
plt.show()
```
